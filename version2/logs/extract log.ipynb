```python
import os
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(6, 4))
for fn in os.listdir("."):
    if(fn.startswith("cls") and fn != "cls_fusion.txt"):
        loss, accuracy = [], []
        with open(fn, "r") as f:
            for i in f.readlines():
                if(i.startswith("Train Epoch")):
                    loss.append(float(i.split(' ')[-1]))
                elif(i.startswith("Vali")):
                    accuracy.append(float(i.split(' ')[-1][1:-3]) / 100)
        plt.plot(list(range(len(loss))), loss, label=fn.split('_')[-1][:-4])
#     elif fn == "cls_fusion.txt":
#         loss, accuracy = [], []
#         with open(fn, "r") as f:
#             for i in f.readlines():
#                 if(i.startswith("Train Epoch")):
#                     loss.append(float(i.split(' ')[-1]))
#                 elif(i.startswith("Vali")):
#                     accuracy.append(float(i.split(' ')[-1][1:-3]) / 100)
#         loss = [loss[_] for _ in range(len(loss)) if _ % 3 == 0]
#         plt.plot(list(range(len(loss))), loss, label=fn.split('_')[-1][:-4])
        
plt.legend()
plt.title("(Classification) Epoch vs. Loss")
plt.show()
```
